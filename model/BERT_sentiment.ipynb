```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
# import requests
# from bs4 import BeautifulStoneSoup
# import re
```

# 2. Instantiate Model

```python
tokenizer = AutoTokenizer.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
model = AutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
```

# 3. Encode and Calculate Sentiment

https://youtu.be/szczpgOEdXs?t=763

```python
# 建立自定義函式庫

def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors = "pt") # pt=pytorch
    result = model(tokens)
    # print(result)
    # print(result.logits)
    sen_score = int(torch.argmax(result.logits))+1
    print(sen_score)
```

```python
# 來試用看看
sentiment_score(df[""].iloc[1])
```

# 4. Collect Reviews

```python
# 利用apply lambda來達成「逐筆生成資料並個別存取」的目的
# 512: 表示tokens的最大值只到512
# https://youtu.be/szczpgOEdXs?t=27:30

df["Sentiment"] = df[""].apply(lambda x : sentiment_score(x[:512]))
```
