```python
import pandas as pd

df = pd.read_csv("/content/drive/Shareddrives/Media Framing/datasets/0_datasets.csv", encoding = "utf-8")
df.head()
```

# 二、實作

```python
test_text = list(df["Headline"])
test_text[2]
```

## 正負面情緒

```python
from cnsenti import Sentiment

senti = Sentiment()
for i in range(len(df)):
    result = senti.sentiment_count(test_text[i])
    print(test_text[i])
    print(result)
    print()
```

## 多種情緒統計

```python
from cnsenti import Emotion
emotion = Emotion()
for i in range(5):
    result = emotion.emotion_count(test_text[i])
    print(test_text[i])
    print(result)
    print()
```
