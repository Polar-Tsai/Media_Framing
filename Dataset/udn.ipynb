# 步驟二、從網頁原始碼觀察是何種網頁類型
然後就可以選擇要import何種函式庫

```python
import json
import requests
```

```python
cate_d = {
    7227 : "sport",
    7225 : "global",
    6639 : "sociaty",
    6641 : "common",
    6644 : "economic",
    6649 : "life",
    6640 : "chinese"
 
}
for i in cate_d.keys():
    print(cate_d[i])
```

```python
# 判斷指定頁數是否有資料
# url = "https://udn.com/api/more?page=5&channelId=2&type=cate_latest_news&cate_id=7227&totalRecNo=231"

def request(url):
    response = requests.get(url)
    posts = json.loads(response.text)
    return posts

# page = 0
# while True:

#     page = page + 1
#     url = "https://udn.com/api/more?page=" + str(page) + "&channelId=2&type=cate_latest_news&cate_id=7227&totalRecNo=231"
#     response = requests.get(url)
#     print(url)
#     if "lists" in request(url).keys():
#         print("yeah")
#     else:
#         break
    
```

```python
table = {
    "Headline" : [],
    "Link" : [],
    "Date" : [],
    "Source" : [],
    "Category" : [],
}
```

```python
page = 0
while True:
    page += 1
    url = "https://udn.com/api/more?page=" + str(page) + "&channelId=2&type=cate_latest_news&cate_id=6639&totalRecNo=231"
    print(url)
    response = requests.get(url)
    # page_xhr = json.loads(response.text)["page"]
    # page_xhr
    if "lists" in request(url).keys():
        posts = json.loads(response.text)["lists"]
        for post in posts:
            headline = post["title"]
            table["Headline"].append(headline)
            print("標題:", headline)
            date = post["time"]
            date = date["date"]
            table["Date"].append(date)
            print("日期:", date)
            link = post["titleLink"]
            table["Link"].append(link)
            print("連結:", link)
            category = url.split("&")[-2]
            category = category.split("=")[-1]
            category = cate_d[int(category)]
            table["Category"].append(category)
            print("類別:", category)
            source = "聯合新聞"
            table["Source"].append(source)
            print("-" * 25)
    else:
        break
```

# 步驟五、把table用pandas顯示

```python
import pandas as pd

# 解決欄位資料有沒抓成功的資料而無法轉成df顯示的問題
df = pd.DataFrame.from_dict(table, orient = "index")
df = df.transpose() # 將df轉置
df.tail(5) # 取倒數5筆資料
```
