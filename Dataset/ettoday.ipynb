# 步驟二、從網頁原始碼觀察是何種網頁類型
-> 報橘，post小秘密+bs4

```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
table = {
    "Headline" : [],
    "Link" : [],
    "Date" : [],
    "Source" : [],
    "Category" : [],
}
```

```python
url = "https://www.ettoday.net/show_roll.php"

page = 0
# while True:
for page in range(145):
    print("頁數:", page + 1)
    # page += 1
    form_data = {
        "offset" : str(page + 1), # 頁數
        "tPage" : 3,
        "tFile" : "20211219.xml",
        "tOt" : 0,
        "tSi" : 100,
        "tAr" : 0
    }

    response = requests.post(url, form_data)
    html = BeautifulSoup(response.text)
    newslists = html.find_all("h3")
    for newslist in newslists:
        headline = newslist.find("a").text
        table["Headline"].append(headline)
        print("標題:", headline)
        date = newslist.find("span", class_ = "date").text
        table["Date"].append(date)
        print("日期:", date)
        category = newslist.find("em", class_ = "tag").text
        table["Category"].append(category)
        print("類別:", category)
        source = "ETtoday"
        table["Source"].append(source)
        link = newslist.find("a")["href"]
        table["Link"].append(link)
        print("連結:", link)
```

# 把table用pandas顯示

```python
import pandas as pd
df = pd.DataFrame(table)
df.info()
```

# 存檔

```python
# 存成csv檔
import os
path = "/content/drive/Shareddrives/Media Framing/datasets/ettoday/"
if not os.path.exists(path):
    os.makedirs(path)

df.to_csv(path + "ettoday.csv", encoding = "utf-8", index = False)
```
