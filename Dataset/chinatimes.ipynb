```python
table = {
    "Headline" : [],
    "Link" : [],
    "Date" : [],
    "Source" : [],
    "Category" : [],
}
```

# 步驟三、先以一頁形式觀察其規律

```python
# newslists = html.find("ul", class_="vertical-list list-style-none").find_all("li")
# # newslists
```

```python
# 解決InsecureRequestWarning: Unverified HTTPS request is being made.
import requests.packages.urllib3
requests.packages.urllib3.disable_warnings()
```

```python
from bs4 import BeautifulSoup
import os
import requests
import pandas as pd


def scrapy(url):
    response = requests.get(url, verify = False)
    response.encoding = "utf-8" # 解決中文亂碼問題
    html = BeautifulSoup(response.text)
    # return html
    newslists = html.find("ul", class_="vertical-list list-style-none").find_all("li")
    for newslist in newslists:
        headline = newslist.find("h3", class_ = "title").find("a").text
        table["Headline"].append(headline)
        print("標題:", headline)
        url = newslist.find("h3").find("a")["href"]
        table["Link"].append(url)
        print("連結:", url)
        date = newslist.find("div", class_ = "meta-info").find("time").find("span", class_ = "date").text
        table["Date"].append(date)
        print("日期:", date)
        source = "中時"
        table["Source"].append(source)
        print("來源:", source)
        category = newslist.find("div", class_ = "meta-info").find("div", class_ = "category").find("a")["href"]
        category = category.split("/")[1]
        # return category
        table["Category"].append(category)
        print("類別:", category)

print(scrapy("https://www.chinatimes.com/politic/total?page=1&chdtv"))
```

# 步驟四、套用步驟三的程式到其他頁數(for-in)，並且append到table

```python
# 解決在頁數不確定的情況下，判斷是否有資料
def request(url):
    response = requests.get(url, verify = False)
    response.encoding = "utf-8" # 解決中文亂碼問題
    html = BeautifulSoup(response.text)
    newslists = html.find("ul", class_="vertical-list list-style-none").find_all("li")
    return newslists


page = 0
while True:
    page = page + 1
    url = "https://www.chinatimes.com/armament/total?page=" + str(page) + "&chdtv"
    # 如果request()沒有資料就停止爬蟲
    if len(request(url)) > 5:
        scrapy(url)
    else:
        break
```

# 步驟五、把table用pandas形式顯示

```python
# 解決欄位資料有沒抓成功的資料而無法轉成df顯示的問題
df = pd.DataFrame.from_dict(table, orient = "index")
df = df.transpose() # 將df轉置
df.tail(5) # 取倒數5筆資料
# df.head()
```
