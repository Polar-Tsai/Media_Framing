```python
from bs4 import BeautifulSoup
import os
import requests
import pandas as pd

def scrapy(url):
    response = requests.get(url, verify = False)
    response.encoding = "utf-8"
    html = BeautifulSoup(response.text)
    newslists = html.find("div", class_ = "row NewsList").find_all("div", class_ = "col-sm-12 newsItems")
    for newslist in newslists:
        headline = newslist.find("h3", class_ = "view-li-title").find("a").text
        table["Headline"].append(headline)
        print("標題:", headline)
        link = newslist.find("h3", class_ = "view-li-title").find("a")["href"]
        table["Link"].append(link)
        print("連結:", link)
        date = newslist.find("div").find("time").text
        table["Date"].append(date)
        print("日期:", date)
        source = "三立"
        table["Source"].append(source)
        print("來源:", source)
        category = newslist.find("div", class_ = "newslabel-tab ").find("a").text
        table["Category"].append(category)
        print("類別:", category)
        print("=" * 25)
# scrapy(url)
```

```python
table = {
    "Headline" : [],
    "Link" : [],
    "Date" : [],
    "Source" : [],
    "Category" : [],
}
```
