```python
import tensorflow
from ckiptagger import WS, POS, NER
# WS = 斷詞, POS = 詞性標註, NER = 實體辨識
```

```python
# 建立NER(實體辨識)的function
text = "台北市,澳門政府今(16)日宣布6月19日起，暫時停止在台辦事處運作，詳細原因並沒有明說。"
def wss(text):
    word_text = ws(text)
    return word_text

def nerr(text):
    word_text = ws(text)
    word_text_pos = pos(word_text)
    word_text_ner = ner(word_text, word_text_pos)
    return word_text_ner
# nerr(text)
# 出問題惹，為什麼會出現一堆空的結果??
wss(text)
```
